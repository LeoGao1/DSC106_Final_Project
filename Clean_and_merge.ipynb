```python
import os
import pandas as pd
import numpy as np
```

```python
directory = './Data'
file_list = []
for filename in os.listdir(directory):
    if filename.endswith(".csv"): 
        temp = os.path.join(directory, filename)
        file_list.append(temp)
        
file_list
```

```python
data = pd.read_csv(file_list[0])
data = data.drop(data.columns[0],axis = 1)

for i,value in enumerate(file_list):
    if(i==0):
        continue
    df = pd.read_csv(file_list[i])
    df = df.drop(df.columns[0],axis = 1)
    data = data.append(df)
    
data.head(10)
```

```python
data
```

```python
file = data
file = file[file.Price != 'Not Priced'] 
file.Name = file.Name.apply(lambda x: 'Delete' if x[0].isalpha() else x)
file = file[file.Name != 'Delete']
file['Year'] = file.Name.apply(lambda x: int(x[0:4]))
file['Name'] = file.Name.apply(lambda x: x[5:])
file['Price'] = file.Price.apply(lambda x: int(x[1:].replace(',','')))
file = file.dropna(subset=['Mileage'])
file['Mileage'] = file.Mileage.apply(lambda x: int(str(x).replace(' mi.','').replace(',','')))
file
```

```python
def change_lower(series):
    series['Name'] = series['Name'].lower()
    return series

file = file.apply(change_lower,axis = 1)
file.head()
```

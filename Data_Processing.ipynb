```python
def mil(x):
    if x < 5000:
        return 0
    elif x < 10000:
        return 5000
    elif x < 15000:
        return 10000
    elif x < 20000:
        return 15000
    elif x < 25000:
        return 20000
    elif x < 30000:
        return 25000
    elif x < 35000:
        return 30000
    elif x < 40000:
        return 35000
    elif x < 45000:
        return 40000
    elif x < 50000:
        return 45000
    elif x < 55000:
        return 50000
    elif x < 60000:
        return 55000
    elif x < 65000:
        return 60000
    elif x < 70000:
        return 65000
    elif x < 75000:
        return 70000
    elif x < 80000:
        return 75000
    elif x < 85000:
        return 80000
    elif x < 90000:
        return 85000
    elif x < 95000:
        return 90000
    elif x < 100000:
        return 95000
    elif x < 105000:
        return 100000
    elif x < 110000:
        return 105000
    elif x < 115000:
        return 110000
    elif x < 120000:
        return 115000
    elif x < 125000:
        return 120000
    else:
        return 125000
```

```python
lst = []
for i in pop10:
    temp = df.loc[df['Name'] == i]
    temp['Mileage'] = temp['Mileage'].apply(mil)
    temp = temp.groupby('Mileage').mean()
    lst.append(list(temp['Current_Price'].sort_values(ascending = False).values))
lst
```

```python
lst = []
def dec_rate(x):
    temp = []
    for i in range(len(x)-1):
        temp.append((x[i]-x[i+1])/x[i])
    return temp
```
